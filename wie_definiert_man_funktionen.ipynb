## Funktionen in Python

```python
def meine_funktion(input_text):
    print("Die Funktion wurde aufgerufen " + input_text)
```

```python
meine_funktion("mit diesem Text als Argument.")
```

```python
def neue_funktion(input_text):
    return "Die Funktion wurde aufgerufen " + input_text
```

```python
txt = "mit diesem Text als Argument."
voller_text = neue_funktion(txt)
print(voller_text)
```

```python
def kugel_volumen(radius, kommastellen):
    pi = 3.14159
    volumen = (4 / 3) * pi * radius ** 3
    volumen = round(volumen, kommastellen)
    
    return pi, radius, volumen

pi, r, vol = kugel_volumen(5, 3)
print("Eine Kugel mit dem Radius " + str(r) + 
        " hat das Volumen " + str(vol) +
        "\nIn der Formel benutzt man pi = " + str(pi))
```
